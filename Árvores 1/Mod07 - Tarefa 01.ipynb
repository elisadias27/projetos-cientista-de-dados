#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}.
````python
dados = pd.get_dummies(dados,columns=['tipo_renda','educacao','estado_civil','tipo_residencia'], drop_first=True)


for column in dados.columns:
    if dados[column].dtype != int and dados[column].dtype != bool and dados[column].dtype != 'uint8':
        dados.drop(columns=column,inplace=True)
    elif dados[column].values.max()>1:
        dados[dados[column] > 1] = dados[dados[column] > 1].replace(dados[column].unique(),1)
dados.rename(columns = {'qtd_filhos':'possui_filhos'},inplace=True)
dados
````
